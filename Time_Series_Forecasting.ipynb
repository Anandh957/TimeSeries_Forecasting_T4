```python
"""
Advanced Time Series Forecasting with Attention-Based Neural Networks
A complete implementation of Encoder-Decoder architecture with self-attention
for multivariate time series forecasting.
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("TensorFlow version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))



# PART 1: DATA GENERATION


class TimeSeriesGenerator:
    """Generate complex multivariate time series with trends, seasonality, and noise."""
    
    def __init__(self, n_timesteps=1000, n_features=5):
        self.n_timesteps = n_timesteps
        self.n_features = n_features
        self.time = np.arange(n_timesteps)
        
    def generate(self):
        """Generate multivariate time series dataset."""
        data = np.zeros((self.n_timesteps, self.n_features))
        
        # Feature 1: Linear trend + annual seasonality + noise
        trend1 = 0.05 * self.time
        season1 = 10 * np.sin(2 * np.pi * self.time / 365)
        noise1 = np.random.normal(0, 2, self.n_timesteps)
        data[:, 0] = trend1 + season1 + noise1
        
        # Feature 2: Quadratic trend + weekly seasonality
        trend2 = 0.0001 * self.time ** 2
        season2 = 5 * np.sin(2 * np.pi * self.time / 7) + 3 * np.cos(2 * np.pi * self.time / 7)
        noise2 = np.random.normal(0, 1.5, self.n_timesteps)
        data[:, 1] = trend2 + season2 + noise2
        
        # Feature 3: Multiple seasonalities (monthly + quarterly)
        season3_1 = 8 * np.sin(2 * np.pi * self.time / 30)
        season3_2 = 4 * np.cos(2 * np.pi * self.time / 90)
        noise3 = np.random.normal(0, 1, self.n_timesteps)
        data[:, 2] = season3_1 + season3_2 + noise3
        
        # Feature 4: Exponential trend + daily pattern
        trend4 = 10 * (1 - np.exp(-self.time / 500))
        season4 = 3 * np.sin(2 * np.pi * self.time / 1)
        noise4 = np.random.normal(0, 0.8, self.n_timesteps)
        data[:, 4] = trend4 + season4 + noise4
        
        # Feature 5: Target variable (complex combination of others)
        data[:, 4] = (0.3 * data[:, 0] + 
                      0.2 * data[:, 1] + 
                      0.25 * data[:, 2] + 
                      0.15 * data[:, 3] +
                      5 * np.sin(2 * np.pi * self.time / 50) +
                      np.random.normal(0, 2, self.n_timesteps))
        
        return data


def create_sequences(data, lookback=50, forecast_horizon=10):
    """Create input-output sequences for training."""
    X, y = [], []
    for i in range(len(data) - lookback - forecast_horizon + 1):
        X.append(data[i:(i + lookback)])
        y.append(data[(i + lookback):(i + lookback + forecast_horizon), -1])
    return np.array(X), np.array(y)



# PART 2: ATTENTION MECHANISM & MODEL ARCHITECTURE


class SelfAttention(layers.Layer):
    """Custom self-attention layer for sequence-to-sequence models."""
    
    def __init__(self, units, **kwargs):
        super(SelfAttention, self).__init__(**kwargs)
        self.units = units
        self.W_query = layers.Dense(units)
        self.W_key = layers.Dense(units)
        self.W_value = layers.Dense(units)
        
    def call(self, inputs):
        """
        inputs: encoder outputs of shape (batch_size, timesteps, features)
        """
        # Compute query, key, value
        query = self.W_query(inputs)
        key = self.W_key(inputs)
        value = self.W_value(inputs)
        
        # Compute attention scores
        scores = tf.matmul(query, key, transpose_b=True)
        scores = scores / tf.math.sqrt(tf.cast(self.units, tf.float32))
        
        # Apply softmax to get attention weights
        attention_weights = tf.nn.softmax(scores, axis=-1)
        
        # Apply attention weights to values
        context = tf.matmul(attention_weights, value)
        
        # Store attention weights for visualization
        self.last_attention_weights = attention_weights
        
        return context, attention_weights
    
    def get_config(self):
        config = super().get_config()
        config.update({"units": self.units})
        return config


class AttentionEncoderDecoder:
    """Encoder-Decoder with Self-Attention for time series forecasting."""
    
    def __init__(self, lookback, n_features, forecast_horizon, 
                 encoder_units=64, decoder_units=64, attention_units=32, dropout=0.2):
        self.lookback = lookback
        self.n_features = n_features
        self.forecast_horizon = forecast_horizon
        self.encoder_units = encoder_units
        self.decoder_units = decoder_units
        self.attention_units = attention_units
        self.dropout = dropout
        self.model = None
        self.attention_model = None
        
    def build_model(self):
        """Build the complete encoder-decoder architecture with attention."""
        
        # Encoder
        encoder_inputs = layers.Input(shape=(self.lookback, self.n_features), name='encoder_input')
        
        # Encoder LSTM layers
        encoder = layers.LSTM(self.encoder_units, return_sequences=True, 
                             dropout=self.dropout, name='encoder_lstm1')(encoder_inputs)
        encoder = layers.LSTM(self.encoder_units, return_sequences=True, 
                             dropout=self.dropout, name='encoder_lstm2')(encoder)
        
        # Self-Attention layer
        attention_layer = SelfAttention(self.attention_units, name='self_attention')
        context, attention_weights = attention_layer(encoder)
        
        # Global average pooling to get fixed-size representation
        context = layers.GlobalAveragePooling1D()(context)
        
        # Decoder
        decoder = layers.RepeatVector(self.forecast_horizon)(context)
        decoder = layers.LSTM(self.decoder_units, return_sequences=True, 
                             dropout=self.dropout, name='decoder_lstm1')(decoder)
        decoder = layers.LSTM(self.decoder_units, return_sequences=True, 
                             dropout=self.dropout, name='decoder_lstm2')(decoder)
        
        # Output layer
        decoder_outputs = layers.TimeDistributed(layers.Dense(1), name='output')(decoder)
        decoder_outputs = layers.Flatten()(decoder_outputs)
        
        # Create model
        self.model = Model(inputs=encoder_inputs, outputs=decoder_outputs, name='AttentionEncoderDecoder')
        
        # Create attention visualization model
        self.attention_model = Model(inputs=encoder_inputs, outputs=attention_weights)
        
        return self.model
    
    def compile_model(self, learning_rate=0.001):
        """Compile the model with optimizer and loss."""
        optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
        self.model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])
        
    def summary(self):
        """Print model summary."""
        if self.model:
            return self.model.summary()
        else:
            print("Model not built yet. Call build_model() first.")



# PART 3: BASELINE MODEL (LSTM without Attention)


def build_baseline_lstm(lookback, n_features, forecast_horizon, units=64, dropout=0.2):
    """Build a baseline LSTM model without attention mechanism."""
    model = keras.Sequential([
        layers.LSTM(units, return_sequences=True, dropout=dropout, 
                   input_shape=(lookback, n_features)),
        layers.LSTM(units, return_sequences=False, dropout=dropout),
        layers.Dense(forecast_horizon)
    ], name='Baseline_LSTM')
    
    model.compile(optimizer=keras.optimizers.Adam(0.001), loss='mse', metrics=['mae'])
    return model



# PART 4: EVALUATION METRICS


def calculate_mase(y_true, y_pred, y_train, seasonality=1):
    """
    Calculate Mean Absolute Scaled Error (MASE).
    
    MASE = MAE / MAE_naive
    where MAE_naive is the MAE of a naive seasonal forecast on training data
    """
    # Calculate MAE of predictions
    mae = np.mean(np.abs(y_true - y_pred))
    
    # Calculate MAE of naive forecast (seasonal random walk)
    naive_error = np.abs(np.diff(y_train, n=seasonality)).mean()
    
    # Avoid division by zero
    if naive_error == 0:
        return np.inf
    
    mase = mae / naive_error
    return mase


def calculate_rmsse(y_true, y_pred, y_train):
    """Calculate Root Mean Squared Scaled Error (RMSSE)."""
    mse = np.mean((y_true - y_pred) ** 2)
    naive_mse = np.mean(np.diff(y_train) ** 2)
    
    if naive_mse == 0:
        return np.inf
    
    rmsse = np.sqrt(mse / naive_mse)
    return rmsse


def evaluate_model(model, X_test, y_test, y_train, model_name="Model"):
    """Comprehensive model evaluation."""
    # Make predictions
    y_pred = model.predict(X_test, verbose=0)
    
    # Flatten if needed
    y_test_flat = y_test.flatten()
    y_pred_flat = y_pred.flatten()
    
    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(y_test_flat, y_pred_flat))
    mae = mean_absolute_error(y_test_flat, y_pred_flat)
    mase = calculate_mase(y_test_flat, y_pred_flat, y_train.flatten())
    rmsse = calculate_rmsse(y_test_flat, y_pred_flat, y_train.flatten())
    
    results = {
        'model': model_name,
        'RMSE': rmse,
        'MAE': mae,
        'MASE': mase,
        'RMSSE': rmsse
    }
    
    return results, y_pred



# PART 5: TRAINING & HYPERPARAMETER TUNING


def rolling_origin_cv(data, model_builder, n_splits=3, lookback=50, 
                      forecast_horizon=10, validation_size=100):
    """
    Rolling origin cross-validation for time series.
    Returns average validation loss across folds.
    """
    scores = []
    
    for i in range(n_splits):
        # Define train/val split for this fold
        train_end = len(data) - validation_size - (n_splits - i - 1) * 50
        val_end = train_end + validation_size
        
        train_data = data[:train_end]
        val_data = data[train_end:val_end]
        
        # Create sequences
        X_train, y_train = create_sequences(train_data, lookback, forecast_horizon)
        X_val, y_val = create_sequences(val_data, lookback, forecast_horizon)
        
        # Build and train model
        model = model_builder()
        history = model.fit(X_train, y_train, 
                          validation_data=(X_val, y_val),
                          epochs=20, batch_size=32, verbose=0)
        
        # Get best validation loss
        best_val_loss = min(history.history['val_loss'])
        scores.append(best_val_loss)
    
    return np.mean(scores)


def hyperparameter_search(data, lookback=50, forecast_horizon=10):
    """
    Systematic hyperparameter search using rolling origin validation.
    """
    param_grid = {
        'learning_rate': [0.001, 0.0005],
        'encoder_units': [64, 128],
        'attention_units': [32, 64],
        'dropout': [0.2, 0.3]
    }
    
    best_score = float('inf')
    best_params = None
    results = []
    
    print("\n" + "="*60)
    print("HYPERPARAMETER SEARCH")
    print("="*60)
    
    # Grid search (simplified for demo - use subset)
    for lr in param_grid['learning_rate']:
        for enc_units in param_grid['encoder_units']:
            for att_units in param_grid['attention_units']:
                for dropout in param_grid['dropout']:
                    
                    params = {
                        'learning_rate': lr,
                        'encoder_units': enc_units,
                        'attention_units': att_units,
                        'dropout': dropout
                    }
                    
                    print(f"\nTesting: {params}")
                    
                    # Define model builder for this configuration
                    def model_builder():
                        att_model = AttentionEncoderDecoder(
                            lookback=lookback,
                            n_features=data.shape[1],
                            forecast_horizon=forecast_horizon,
                            encoder_units=enc_units,
                            decoder_units=enc_units,
                            attention_units=att_units,
                            dropout=dropout
                        )
                        att_model.build_model()
                        att_model.compile_model(learning_rate=lr)
                        return att_model.model
                    
                    # Evaluate using rolling origin CV
                    score = rolling_origin_cv(data, model_builder, n_splits=2,
                                            lookback=lookback, 
                                            forecast_horizon=forecast_horizon)
                    
                    results.append({**params, 'cv_score': score})
                    print(f"CV Score: {score:.4f}")
                    
                    if score < best_score:
                        best_score = score
                        best_params = params
    
    print("\n" + "="*60)
    print("BEST PARAMETERS:")
    print(f"Parameters: {best_params}")
    print(f"CV Score: {best_score:.4f}")
    print("="*60 + "\n")
    
    return best_params, results



# PART 6: VISUALIZATION


def plot_predictions(y_true, y_pred, n_windows=3, forecast_horizon=10):
    """Visualize predictions vs actual for multiple test windows."""
    fig, axes = plt.subplots(n_windows, 1, figsize=(15, 4*n_windows))
    
    if n_windows == 1:
        axes = [axes]
    
    for i in range(n_windows):
        start_idx = i * forecast_horizon * 2
        end_idx = start_idx + forecast_horizon * 3
        
        axes[i].plot(range(start_idx, end_idx), 
                    y_true.flatten()[start_idx:end_idx], 
                    label='Actual', marker='o', linewidth=2)
        axes[i].plot(range(start_idx, end_idx), 
                    y_pred.flatten()[start_idx:end_idx], 
                    label='Predicted', marker='s', linewidth=2, alpha=0.7)
        axes[i].set_title(f'Test Window {i+1}')
        axes[i].set_xlabel('Time Step')
        axes[i].set_ylabel('Value')
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)
    
    plt.tight_layout()
    return fig


def visualize_attention_weights(attention_model, X_sample, n_samples=3):
    """Extract and visualize attention weights."""
    fig, axes = plt.subplots(n_samples, 1, figsize=(15, 4*n_samples))
    
    if n_samples == 1:
        axes = [axes]
    
    for i in range(n_samples):
        # Get attention weights for sample
        sample = X_sample[i:i+1]
        attention_weights = attention_model.predict(sample, verbose=0)[0]
        
        # Plot heatmap
        sns.heatmap(attention_weights, cmap='viridis', ax=axes[i], 
                   cbar_kws={'label': 'Attention Weight'})
        axes[i].set_title(f'Attention Weights - Sample {i+1}')
        axes[i].set_xlabel('Key Position (Time Step)')
        axes[i].set_ylabel('Query Position (Time Step)')
    
    plt.tight_layout()
    return fig



# MAIN EXECUTION


def main():
    """Main execution pipeline."""
    
    print("\n" + "="*60)
    print("ADVANCED TIME SERIES FORECASTING WITH ATTENTION")
    print("="*60 + "\n")
    
    # Parameters
    LOOKBACK = 50
    FORECAST_HORIZON = 10
    TRAIN_SIZE = 0.7
    VAL_SIZE = 0.15
    
    # Step 1: Generate Data
    print("Step 1: Generating multivariate time series data...")
    generator = TimeSeriesGenerator(n_timesteps=1000, n_features=5)
    data = generator.generate()
    print(f"Generated data shape: {data.shape}")
    
    # Normalize data
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)
    
    # Create sequences
    X, y = create_sequences(data_scaled, LOOKBACK, FORECAST_HORIZON)
    print(f"Sequences shape - X: {X.shape}, y: {y.shape}")
    
    # Split data
    train_idx = int(len(X) * TRAIN_SIZE)
    val_idx = int(len(X) * (TRAIN_SIZE + VAL_SIZE))
    
    X_train, y_train = X[:train_idx], y[:train_idx]
    X_val, y_val = X[train_idx:val_idx], y[train_idx:val_idx]
    X_test, y_test = X[val_idx:], y[val_idx:]
    
    print(f"Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")
    
    # Step 2: Hyperparameter Search (simplified)
    print("\nStep 2: Conducting hyperparameter search...")
    best_params, hp_results = hyperparameter_search(
        data_scaled, LOOKBACK, FORECAST_HORIZON
    )
    
    # Step 3: Build and Train Final Model
    print("\nStep 3: Building and training final attention model...")
    final_model = AttentionEncoderDecoder(
        lookback=LOOKBACK,
        n_features=data.shape[1],
        forecast_horizon=FORECAST_HORIZON,
        encoder_units=best_params['encoder_units'],
        decoder_units=best_params['encoder_units'],
        attention_units=best_params['attention_units'],
        dropout=best_params['dropout']
    )
    
    final_model.build_model()
    final_model.compile_model(learning_rate=best_params['learning_rate'])
    print("\nModel Architecture:")
    final_model.summary()
    
    # Callbacks
    early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
    
    # Train
    history = final_model.model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=100,
        batch_size=32,
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )
    
    # Step 4: Build and Train Baseline
    print("\nStep 4: Training baseline LSTM model...")
    baseline_model = build_baseline_lstm(
        LOOKBACK, data.shape[1], FORECAST_HORIZON
    )
    
    baseline_history = baseline_model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=100,
        batch_size=32,
        callbacks=[early_stop],
        verbose=0
    )
    
    # Step 5: Evaluate Models
    print("\nStep 5: Evaluating models...")
    
    attention_results, attention_pred = evaluate_model(
        final_model.model, X_test, y_test, y_train, "Attention Encoder-Decoder"
    )
    
    baseline_results, baseline_pred = evaluate_model(
        baseline_model, X_test, y_test, y_train, "Baseline LSTM"
    )
    
    # Print results
    print("\n" + "="*60)
    print("FINAL MODEL PERFORMANCE")
    print("="*60)
    print("\nAttention Encoder-Decoder:")
    for metric, value in attention_results.items():
        if metric != 'model':
            print(f"  {metric}: {value:.4f}")
    
    print("\nBaseline LSTM:")
    for metric, value in baseline_results.items():
        if metric != 'model':
            print(f"  {metric}: {value:.4f}")
    
    # Calculate improvement
    print("\n" + "="*60)
    print("PERFORMANCE IMPROVEMENT")
    print("="*60)
    for metric in ['RMSE', 'MAE', 'MASE']:
        improvement = ((baseline_results[metric] - attention_results[metric]) / 
                      baseline_results[metric] * 100)
        print(f"{metric} Improvement: {improvement:.2f}%")
    
    # Step 6: Visualizations
    print("\nStep 6: Generating visualizations...")
    
    # Training history
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Attention Model Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    plt.plot(baseline_history.history['loss'], label='Train Loss')
    plt.plot(baseline_history.history['val_loss'], label='Val Loss')
    plt.title('Baseline LSTM Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('training_history.png', dpi=150, bbox_inches='tight')
    print("Saved: training_history.png")
    
    # Predictions
    pred_fig = plot_predictions(y_test, attention_pred, n_windows=3, 
                                forecast_horizon=FORECAST_HORIZON)
    pred_fig.savefig('predictions_comparison.png', dpi=150, bbox_inches='tight')
    print("Saved: predictions_comparison.png")
    
    # Attention weights
    att_fig = visualize_attention_weights(final_model.attention_model, X_test, n_samples=3)
    att_fig.savefig('attention_weights.png', dpi=150, bbox_inches='tight')
    print("Saved: attention_weights.png")
    
    print("\n" + "="*60)
    print("EXECUTION COMPLETE")
    print("="*60 + "\n")
    
    return {
        'attention_model': final_model,
        'baseline_model': baseline_model,
        'attention_results': attention_results,
        'baseline_results': baseline_results,
        'history': history,
        'data': data_scaled,
        'X_test': X_test,
        'y_test': y_test
    }


if __name__ == "__main__":
    results = main()
```
